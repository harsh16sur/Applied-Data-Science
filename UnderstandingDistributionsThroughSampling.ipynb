```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

# def 

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)

# plot the histograms
f,axes=plt.subplots(1,4,sharey=True,figsize=(9,3))
axes[0].hist(x1, normed=True, bins=20, alpha=0.5,color='blue')
axes[1].hist(x2, normed=True, bins=20, alpha=0.5,color='red')
axes[2].hist(x3, normed=True, bins=20, alpha=0.5,color='yellow')
axes[3].hist(x4, normed=True, bins=20, alpha=0.5,color='orange');
# plt.axis([-7,21,0,0.6])
axes[0].set_title('Normal')
axes[1].set_title('Gamma')
axes[2].set_title('Exponential')
axes[3].set_title('Uniform')

axes[0].set_xlabel('x')
axes[1].set_xlabel('x')
axes[2].set_xlabel('x')
axes[3].set_xlabel('x')

axes[0].set_ylabel('y')
axes[1].set_ylabel('y')
axes[2].set_ylabel('y')
axes[3].set_ylabel('y')
plt.annotate('n=dist',[3,27])
plt.subplots_adjust(bottom=0.25)


# plt.text(x1.mean()-1.5, 0.5, 'x1\nNormal')
# plt.text(x2.mean()-1.5, 0.5, 'x2\nGamma')
# plt.text(x3.mean()-1.5, 0.5, 'x3\nExponential')
# plt.text(x4.mean()-1.5, 0.5, 'x4\nUniform')
```
